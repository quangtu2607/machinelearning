```
#PROBLEM1
import matplotlib.pyplot as plt
rice = 1
lst = [rice]
for i in range(0,99):
  rice *= 2
  lst.append(rice)
print("Total grains of rice on 100th day: {}".format(rice))

%matplotlib inline
plt.title("Number of grains of rice on the 100th day")
plt.xlabel("Number of days")
plt.ylabel("Grains of rice")
plt.plot(lst, color='green', marker='x', linestyle='dashed', linewidth=1)
plt.show()
```

```
#PROBLEM2
import matplotlib.pyplot as plt
def compute_sorori_shinzaemon(day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    # Write the code here. pass is an instruction that means do nothing.
    rice = 1
    list_n_grains = [rice]
    for i in range(0, day - 1):
      rice *= 2
      list_n_grains.append(rice)
    list_total_grains = rice
    return list_n_grains, list_total_grains
day = int(input("Enter number of days: "))
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day)
print("Total grains of rice: {}".format(list_total_grains))
%matplotlib inline
plt.title("Number of grains of rice on the 100th day")
plt.xlabel("Number of days")
plt.ylabel("Grains of rice")
plt.plot(list_n_grains, color='green', marker='x', linestyle='dashed', linewidth=1)
plt.show()
```

```
#PROBLEM3
def people_v_rice(n_people, n_grains_of_rice):
  total_consumption_p_day = n_people * 68.5
  n_days_to_live = n_grains_of_rice / total_consumption_p_day
  return n_days_to_live
n_people = int(input("Number of people: "))
n_grains_of_rice = int(input("Total grains of rice: "))
n_days_to_live = people_v_rice(n_people, n_grains_of_rice)
print("Number of days: {} day(s)".format(n_days_to_live))
```
