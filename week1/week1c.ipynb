The formula for the volume of the sphere is V = 4/3 πr³. Assume that the radius of the chestnut bun is 5 cm. Voyager 1, the most distant human-made object, has been in space for more than 40 years and it still has not escaped the influence of our Sun. Voyager 1 is about 13.8 billion miles (22.2 billion kilometers) from the Sun. Therefore we take 22.2 billion km as the radius of the solar system.

```
import math

r_chestnut = 0.05
r_solar = 22.2 * 10**9
V_chestnut = 4 * math.pi * r_chestnut**3 / 3
V_solar = 4 * math.pi * r_solar**3 / 3
V_chestnut_total = V_chestnut
t = 0
while V_chestnut_total < V_solar:
  V_chestnut_total *= 2
  t += 5
print("Number of times that chestnut buns cover the solar system: {} minute(s)".format(t))

```

Assume that radius of the ball is 10 cm and Tokyo Dome is 10 m.

```
#generic program
import math
import matplotlib.pyplot as plt

r1 = float(input("Radius of object sprinkled with Byvine: "))
r2 = float(input("Radius of the container: "))
V1 = 4 * math.pi * r1**3 / 3
V2 = 4 * math.pi * r2**3 / 3
V1_total = V1
t = 0
lst1 = [t]
lst2 = [V1_total] 
while V1_total < V2:
  V1_total *= 2
  t += 5
  lst1.append(t)
  lst2.append(V1_total)
print("Number of times for an object sprinkled by Byvine to cover the container: {} minute(s)".format(t))

%matplotlib inline
plt.title("")
plt.xlabel("Volume has been covered")
plt.ylabel("Times ")
plt.plot(lst2, lst1, marker = 'x', linestyle='dashed', linewidth=1)
plt.show()
```
