```
#PROBLEM2
THICKNESS = 0.00008
folded_paper = THICKNESS
distance = 0
n = 0
def n_of_diff(distance, n, folded_paper):
  while folded_paper < distance:
    folded_paper *= 2
    n += 1
  return n
distance = 4.2465 * 300000 * 365 * 24 * 60 * 60
N = n_of_diff(distance, n, folded_paper)
print("Number of folds: {}".format(N))
```

```
#PROBLEM3
import math

THICKNESS = 0.00008
folded_paper = THICKNESS
distance = 0
n = 0
def length_of_paper(distance, n, folded_paper):
  def n_of_diff(distance, n, folded_paper):
    while folded_paper < distance:
      folded_paper *= 2
      n += 1
    return n
  n = n_of_diff(distance, n, folded_paper)
  L = math.pi * THICKNESS * (2**n + 4) * (2**n - 1) / 6
  return L

distance = 3776
L = length_of_paper(distance, n, folded_paper)
print("Length of paper to reach Mount Fuji: {}".format(L))

distance = 384400000
L = length_of_paper(distance, n, folded_paper)
print("Length of paper to reach the Moon: {}".format(L))

distance = 4.2465 * 300000 * 365 * 24 * 60 * 60
L = length_of_paper(distance, n, folded_paper)
print("Length of paper to reach the stars closest to the Sun: {}".format(L))
  
```
