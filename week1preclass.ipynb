```
"""
Code to calculate the thickness when the paper is folded once
"""
THICKNESS = 0.00008
folded_thickness = THICKNESS*2**43
print("Thickness: {} meters".format(folded_thickness))
```

```
# Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
```

```
THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(0,43):
  folded_thickness *= 2

print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
```

```
%%timeit
import time
start = time.time()
#####
# Paste the code you want to compare here
THICKNESS = 0.00008
folded_thickness = THICKNESS
for i in range(0,43):
  folded_thickness *= 2

print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```
THICKNESS = 0.00008
folded_thickness = THICKNESS
lst = [folded_thickness]
for i in range(0,43):
  folded_thickness *= 2
  lst.append(folded_thickness)
print(lst)
```

```
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(lst) # Enter the variable name of the list in "List name"
plt.show()
```

```
"""
Display a red line graph.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(lst, color='red')
plt.show()
```

```
"""
Display the axis value in large size.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=20) # Make settings related to axis values
plt.plot(lst)
plt.show()
```
