```python
#PROBLEM5
import numpy as np
import matplotlib.pyplot as plt

csv_path = "/mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)

ele = [i[3] for i in fuji]
pos = [i[0] for i in fuji]

point_list = np.column_stack((pos, ele))

def slope_of_a_point(current_point_num, list_of_point): 
  dY = list_of_point[current_point_num][1] - list_of_point[current_point_num - 1][1]
  dX = list_of_point[current_point_num][0] - list_of_point[current_point_num - 1][0]
  sl = dY/dX
  return sl

def destination_point(curr_point, hyperpara, curr_point_gra):
  try:
    dest_point = curr_point - hyperpara*curr_point_gra 
    dest_point = round(dest_point)
    return dest_point
  except dest_point < 0:
    print("Error")

def go_down(start_point, point_list):
  curr_point = start_point
  dest_point = -1
  point_track = [start_point]
  while curr_point != dest_point:
    if dest_point != -1:
      curr_point = dest_point
    slope = slope_of_a_point(curr_point, point_list)
    dest_point = destination_point(curr_point, 0.1, slope)
    point_track.append(dest_point)
  return point_track
  
point_track = go_down(136, point_list)
ele_at_pos = []
for point in point_track:
  for i in fuji:
    if point == i[0]:
      ele_at_pos.append(i[3])

plt.title('Points moved in cross section')
plt.xlabel('position')
plt.ylabel('elevation(m)')
plt.plot(pos, ele)
plt.plot(point_track, ele_at_pos, marker='.')
plt.show()

plt.title('Line graph of the altitude and slope')
plt.xlabel('position')
plt.ylabel('elevation(m)')
plt.plot(point_track, ele_at_pos, marker='.')
plt.show()
```
