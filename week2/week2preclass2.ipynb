```
#PROBLEM2
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x1,x2 = np.random.multivariate_normal([-3, 0], [[1, 0.8], [0.8, 1]], 500).T

%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("sample")
plt.scatter(x1, x2) 
plt.show()
```

```
#PROBLEM3
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x1, x2 = np.random.multivariate_normal([-3, 0], [[1, 0.8], [0.8, 1]], 500).T

%matplotlib inline
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.title("sample")
plt.xlim(-5, 2)
plt.hist(x1, 50, density=True)
plt.show()

%matplotlib inline
plt.xlabel("x2")
plt.ylabel("Frequency")
plt.title("sample")
plt.xlim(-5, 2)
plt.hist(x2, 50, density=True)
plt.show()
```

```
#PROBLEM4
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x3, x4 = np.random.multivariate_normal([-3, 0], [[1, 0.8], [0.8, 1]], 500).T
x1, x2 = np.random.multivariate_normal([0, -3], [[1, 0.8], [0.8, 1]], 500).T


%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("sample")
t1 = plt.scatter(x3, x4)
t2 = plt.scatter(x1, x2)
plt.legend((t1, t2), ('0', '1'))
plt.show()
```

```
#PROBLEM5
import numpy as np

np.random.seed(0)
a1 = np.random.multivariate_normal([-3, 0], [[1, 0.8], [0.8, 1]], 500)
a2 = np.random.multivariate_normal([0, -3], [[1, 0.8], [0.8, 1]], 500)

np.concatenate((a1, a2))
```

```
#PROBLEM6
import numpy as np

np.random.seed(0)
a1 = np.random.multivariate_normal([-3, 0], [[1, 0.8], [0.8, 1]], 500)
source = a1.tolist()
a1 = [row + [0] for row in source]

a2 = np.random.multivariate_normal([0, -3], [[1, 0.8], [0.8, 1]], 500)
source = a2.tolist()
a2 = [row + [1] for row in source]

np.concatenate((a1, a2))
```
