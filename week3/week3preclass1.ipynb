The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in 1936. Two of the three species were collected in the Gaspé Peninsula, and all of them are from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus. The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.

```python
#PROBLEM6b
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
X = pd.DataFrame({'sepal_length': data.data[:,0], 'sepal_width': data.data[:,1], 'petal_length': data.data[:,2], 'petal_width': data.data[:,3]})
y =  pd.DataFrame({'Species': data.target})

frames = [X, y]
df = pd.concat(frames, axis=1)

def drawBoxPlot(data, ax, color):
  sns.boxplot(data=data, orient='v', color = color, ax = ax)

def drawViolinPlot(data, ax, color):
  sns.violinplot(data=data, orient='h', color = color, ax = ax)

fig, axes = plt.subplots(2,2, figsize=(10,8))
fig.suptitle('box plot')
drawBoxPlot(data=df.loc[:,'sepal_length'],  ax=axes[0, 0], color='red')
drawBoxPlot(data=df.loc[:,'sepal_width'],   ax=axes[0, 1], color='blue')
drawBoxPlot(data=df.loc[:,'petal_length'],  ax=axes[1, 0], color='green')
drawBoxPlot(data=df.loc[:,'petal_width'],   ax=axes[1, 1], color='yellow')
```

```python
#PROBLEM7a
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
from math import floor

data = load_iris()
X = pd.DataFrame({'sepal_length': data.data[:,0], 'sepal_width': data.data[:,1], 'petal_length': data.data[:,2], 'petal_width': data.data[:,3]})
y =  pd.DataFrame({'Species': data.target})

frames = [X, y]
df = pd.concat(frames, axis=1)

names = ['sepal_length', 'sepal_width','petal_length','petal_width']
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
pos = 0
for i in names:
  for j in names:
    if i == j: break
    sns.scatterplot(data=df, x=i, y=j, ax=axes[floor(pos/3), pos % 3], hue='Species')
    pos = pos + 1
```
